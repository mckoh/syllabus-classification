```python
from langchain_community.vectorstores import SKLearnVectorStore
from langchain_ollama import ChatOllama, OllamaEmbeddings
from langchain_core.prompts.prompt import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_community.document_loaders import TextLoader, JSONLoader
from os import listdir
from os.path import join
from itertools import chain
```

```python
prompt = PromptTemplate(
    template="""You are an assistant for question-answering tasks.
    Use the following documents to answer the question.
    If you don't know the answer, just say that you don't know.
    Question: {question}
    Documents: {documents}
    Answer:
    """,
    input_variables=["question", "documents"],
)
```

```python
folder = "input/FHKu_DSIA"
files = [TextLoader(join(folder, file)).load() for file in listdir(folder) if file.endswith(".json")]
files = [item for sublist in files for item in sublist]

vectorstore = SKLearnVectorStore.from_documents(
    documents=files,
    embedding=OllamaEmbeddings(model="qwen3-embedding:8b"),
)
```

```python
retriever = vectorstore.as_retriever()
```

```python
retriever.invoke("Visualisierung von Daten")
```
