# Automatic Syllabus Classification

```python
from os import listdir
from os.path import join
from datetime import datetime as dt

from syllabus_classificaiton.functions import load_curriculum
from syllabus_classificaiton.functions import transform_curriculum
from syllabus_classificaiton.functions import classify_curriculum
from syllabus_classificaiton.functions import save_classification
```

```python
start_time = dt.now()
in_folder = "input/FHKu_DSIA"
out_folder = "output/FHKu_DSIA"

t = len(listdir(in_folder))

for i, file in enumerate(listdir(in_folder)):
    if file.endswith(".json"):
        print(i+1, "von", t)
        src_file = file
        dst_file = file.split(".")[0]+".md"

        in_path = join(in_folder, src_file)
        out_path = join(out_folder, dst_file)

        x = load_curriculum(file_path=in_path, verbose=0)
        x = transform_curriculum(json_curriculum=x, verbose=0)
        c = classify_curriculum(txt_curriculum=x, verbose=0)
        s = save_classification(output_path=out_path, txt_classification=c, verbose=0)

print("TOTAL TIME:", dt.now()-start_time)
```
