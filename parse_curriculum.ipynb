# Parse DSIA Curriculum

```python
from requests import get
from bs4 import BeautifulSoup
from string import punctuation
from time import sleep
from os.path import join
from json import dumps
```

```python
base_url = "https://www.fh-kufstein.ac.at"
overview = base_url + "/master/data-science-intelligent-analytics-bb"

response = get(overview)
content = response.content
soup = BeautifulSoup(content, "html.parser")
lvs = soup.find_all("button", attrs={"class": "curriculum-course-row-head"})

for lv in lvs:
    name = lv.find_next("h4").text
    url = lv.find_next("a")["href"]
    name = "".join([char for char in name if char not in punctuation]).lower().replace("  ", " ").replace(" ", "_")

    sub_url = base_url+url
    sub_soup = BeautifulSoup(get(sub_url).content, "html.parser")

    sections = sub_soup.find_all("div", attrs={"class": "section__content"})

    lv_content = {}

    for section in sections:
        header = section.find_next("h3").text
        body = section.find_next("div", attrs={"class": "section__content-body wysiwyg-ul"}).text
        body = body.strip()

        lv_content[header] = body

    with open(join("input", name+".json"), "w", encoding="utf-8") as file:
        file.write(dumps(lv_content))

    sleep(3)
```
